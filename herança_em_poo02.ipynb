### Exercício 3: Construtor `super()`

**Objetivo**: Crie uma classe `FormaGeometrica` com `cor`. Crie uma classe `Circulo` que herda de `FormaGeometrica`, adiciona `raio` e utiliza `super()` para inicializar a cor.

```python
class FormaGeometrica():
  def __init__(self, cor):
    self.cor = cor

class Circulo(FormaGeometrica):
  def __init__(self, cor, raio):
    super().__init__(cor)
    self.raio = raio

  def info(self):
    print(f"raio : {self.raio}")
    print(f"cor: {self.cor}")

forma01 = Circulo("azul", 23)
forma01.info()
```

### Exercício 4: Herança Hierárquica

**Objetivo**: Crie uma classe `Empregado`. A partir dela, crie `Gerente` e `Desenvolvedor`, cada um com atributos e métodos específicos do seu cargo.

```python
class Empregado():
  def __init__(self, nome, idade):
    self.nome = nome
    self.idade = idade

  def info(self):
    print(f"nome: {self.nome}")
    print(f"idade: {self.idade}")

class Gerente(Empregado):
  def __init__(self, nome, idade, saldo_empresarial):
    super().__init__(nome, idade)
    self.saldo_empresarial = saldo_empresarial

  def info_gerente(self):
    print(f"nome: {self.nome}")
    print(f"idade: {self.idade}")
    print(f"saldo empresarial:{self.saldo_empresarial}")
    print("-=" * 30)
class Desenvolvedor(Empregado):
  def __init__(self, nome, idade, linguagem_de_programacao):
    super().__init__(nome, idade)
    self.linguagem_de_programacao = linguagem_de_programacao

  def info_desenvolvedor(self):
    print(f"mome:{self.nome}")
    print(f"idade: {self.idade}")
    print(f"linguagem de programacao: {self.linguagem_de_programacao}")

emp1 = Gerente("anima", 23, 3000)
emp2 =Desenvolvedor("erik", 14, "python")
emp1.info_gerente()
emp2.info_desenvolvedor()
```

### Exercício 5: Herança Multinível

**Objetivo**: Crie uma classe `Veiculo`. Deriva `Carro` de `Veiculo`, e `CarroEsportivo` de `Carro`, demonstrando uma cadeia de herança.

```python
class Veiculo():
  def __init__(self, marca, modelo):
    self.marca = marca
    self.modelo = modelo
  def info(self):
    print(f" marca: {self.marca}")
    print(f"modelo:{self.modelo}")

class Carro(Veiculo):
  def __init__(self, marca, modelo, velocidade):
    super().__init__(marca, modelo)
    self.velocidade = velocidade

  def info(self):
    print(f"marca: {self.marca}")
    print(f"modelo: {self.modelo}")
    print(f"velocidade: {self.velocidade}")
    print("-=" * 30)

class CarroEsportivo(Carro):
  def __init__(self, marca, modelo, velocidade):
    super().__init__(marca, modelo, velocidade)

  def info(self):
    print(f"marca: {self.marca}")
    print(f'modelo: {self.modelo}')

  def modo_esportivo(self):
    print("modo esportivo ativado")


car01 = Carro("volvo", 2011, 250)
car02 = CarroEsportivo("fusion", 2011, 280)
car01.info()
car02.info()
car02.modo_esportivo()
```

### Exercício 6: Herança Múltipla (Mixins)

**Objetivo**: Crie classes `Nadador` e `Corredor` (mixins). Crie uma classe `Triatleta` que herda de ambas e possui habilidades de nadar e correr.

```python
class Nadador():
  def __init__(self, nome, velocidade_de_natacao):
    self.nome = nome
    self.velocidade_de_natacao = velocidade_de_natacao

  def nadar(self):
    print("estou nadando")

class Corredor():
  def __init__(self, nome, velocidade, distancia_percorrida):
    self.nome = nome
    self.velocidade= velocidade
    self.distancia_percorrida = distancia_percorrida

  def correr(self):
    print("estou correndo")


class Triatleta(Nadador, Corredor):
  def __init__(self, nome, velocidade, distancia_percorrida, velocidade_de_natacao):
    Nadador.__init__(self, nome, velocidade_de_natacao)
    Corredor.__init__(self, nome, velocidade, distancia_percorrida)

  def Triatleta(self):
    print("parabens")


p0 = Nadador("erik", 32)
p1 = Triatleta("erik", 32, 21, 12)
p1.Triatleta()
```
