````python
from abc import ABC, abstractmethod

class FormGeometric(ABC):
  @abstractmethod
  def calculate_area(self):
    pass

  @abstractmethod
  def calculate_perimeter(self):
    pass


class Rectangle(FormGeometric):
  def calculate_area(self, base, height):
    result = base * height /2
    return f"the result of the area is: {result}"
    print()
  def calculate_perimeter(self, largura, height):
    result = 2* (largura * height)
    return f"the result of the perimeter is: {result}"
    print()

class Circle(FormGeometric):
  def calculate_area(self, radius):
    result = 3.14* (radius * radius)
    return f" the result of the area in the circle is: {result}"
    print()
  def calculate_perimeter(self, radius):
    radius = radius * radius
    diameter = 2* radius
    result = 3.14 * diameter
    return f"the resulte of the perimeter in the circle is: {result}"
    print()

fig01 = Rectangle()
print(fig01.calculate_area(22, 34))
print(fig01.calculate_perimeter(45,34))
print()

fig02 = Circle()
print(fig02.calculate_area(45))
print(fig02.calculate_perimeter(45))


````

### Exercício 2: Classe Abstrata `Animal`

Crie uma classe abstrata chamada `Animal` com os seguintes métodos abstratos:

*   `emitir_som()`: Imprime o som que o animal faz.
*   `mover()`: Imprime como o animal se move.

Implemente duas classes concretas, por exemplo, `Cachorro` e `Passaro`, que herdem de `Animal` e forneçam suas próprias implementações para esses métodos.

````python
from abc import ABC, abstractmethod

class Animal(ABC):
  @abstractmethod
  def emitir_som(self):
    pass
  @abstractmethod
  def mover(self):
    pass

class Cachorro(Animal):
  def emitir_som(self):
     print(" au au")

  def mover(self):
    print("o cachorro esta se movendo")

class Passaro(Animal):
  def emitir_som(self):
    print("piu piu")

  def mover(self):
    print("o passaro esta voando")

animal01 = Cachorro()
print(animal01.emitir_som())
print(animal01.mover())
print()

animal02 = Passaro()
print(animal02.emitir_som())
print(animal02.mover())
print()
print("fim")
````

### Exercício 3: Classe Abstrata `DispositivoEletronico`

Defina uma classe abstrata `DispositivoEletronico` com os métodos abstratos:

*   `ligar()`: Imprime que o dispositivo está ligando.
*   `desligar()`: Imprime que o dispositivo está desligando.

Crie classes concretas como `Televisao` e `Smartphone` que herdem e implementem esses comportamentos. Você pode adicionar métodos específicos para cada dispositivo (e.g., `mudar_canal` para `Televisao`, `fazer_chamada` para `Smartphone`).

````python
from abc import ABC, abstractmethod
class DispositivoEletronico(ABC):
  @abstractmethod
  def ligar(self):
    print("o dispositivo esta ligado")

  @abstractmethod
  def desligar(self):
    print("dipositivo desligando")

class Televisao(DispositivoEletronico):
  def ligar(self):
     return super().ligar()

  def desligar(self):
     return super().desligar()

  def mudar_canal(self):
    print("mudando de canal")

class Smartphone(DispositivoEletronico):
  def ligar(self):
     return super().ligar()

  def desligar(self):
     return super().desligar()

  def fazer_chamada(self):
    print("fazendo a chamada")


dispositivo01 = Televisao()
print(dispositivo01.ligar())
print(dispositivo01.mudar_canal())
print(dispositivo01.desligar())
print()

dispositivo02 = Smartphone()
print(dispositivo02.ligar())
print(dispositivo02.fazer_chamada())
print(dispositivo02.desligar())
````

````python
from abc import ABC, abstractmethod

# Definindo uma classe abstrata 'Veiculo'
class Veiculo(ABC):
    @abstractmethod
    def ligar(self):
        pass

    @abstractmethod
    def desligar(self):
        pass

    def abastecer(self):
        print("Abastecendo o veículo.")

# Classe concreta 'Carro' que herda de Veiculo
class Carro(Veiculo):
    def ligar(self):
        print("Carro ligado: motor está funcionando.")

    def desligar(self):
        print("Carro desligado: motor está parado.")

    def abrir_porta(self):
        print("Abrindo a porta do carro.")

# Classe concreta 'Moto' que herda de Veiculo
class Moto(Veiculo):
    def ligar(self):
        print("Moto ligada: ignição acionada.")

    def desligar(self):
        print("Moto desligada: motor parou de girar.")

    def empinar(self):
        print("Empinando a moto!")

# Tentativa de instanciar a classe abstrata (irá gerar um erro)
# meu_veiculo = Veiculo()

# Instanciando classes concretas
meu_carro = Carro()
minha_moto = Moto()


print("\n--- Operações com o Carro ---")
meu_carro.ligar()
meu_carro.abastecer()
meu_carro.abrir_porta()
meu_carro.desligar()

print("\n--- Operações com a Moto ---")
minha_moto.ligar()
minha_moto.abastecer()
minha_moto.empinar()
minha_moto.desligar()
````
