```python
from abc import ABC, abstractmethod

class Produto(ABC):
  def __init__(self, nome, preco_base, quantidade_estoque):
    self._nome = nome  # Atributo protegido
    self._preco_base = preco_base  # Atributo protegido
    self._quantidade_estoque = quantidade_estoque  # Atributo protegido

  # Getters
  def get_nome(self):
    return self._nome

  def get_preco_base(self):
    return self._preco_base

  def get_quantidade_estoque(self):
    return self._quantidade_estoque

  # Setter para quantidade_estoque com validação
  def set_quantidade_estoque(self, nova_quantidade):
    if nova_quantidade >= 0:
      self._quantidade_estoque = nova_quantidade
      print(f"Quantidade de estoque para '{self._nome}' atualizada para {self._quantidade_estoque}.")
    else:
      print("Erro: A quantidade de estoque não pode ser negativa.")

  @abstractmethod
  def calcular_preco_final(self):
    pass

  @abstractmethod
  def exibir_detalhes(self):
    pass


class Eletronico(Produto):
  def __init__(self, nome, preco_base, quantidade_estoque, garantia_meses):
    super().__init__(nome, preco_base, quantidade_estoque)
    self._garantia_meses = garantia_meses  # Atributo específico protegido

  def calcular_preco_final(self):
    # Se a garantia for maior que 12 meses, adiciona 15%; caso contrário, adiciona 10%.
    if self._garantia_meses > 12:
      return self._preco_base * 1.15
    else:
      return self._preco_base * 1.10

  def exibir_detalhes(self):
    print("\n" + "-=" * 20)
    print("Detalhes do Produto Eletrônico:")
    print(f"  Nome: {self.get_nome()}")
    print(f"  Preço Base: R${self.get_preco_base():.2f}")
    print(f"  Garantia: {self._garantia_meses} meses")
    print(f"  Quantidade em Estoque: {self.get_quantidade_estoque()}")
    print(f"  Preço Final Calculado: R${self.calcular_preco_final():.2f}")
    print("-=" * 20)


class Livro(Produto):
  def __init__(self, nome, preco_base, quantidade_estoque, autor, paginas):
    super().__init__(nome, preco_base, quantidade_estoque)
    self._autor = autor  # Atributo específico protegido
    self._paginas = paginas  # Atributo específico protegido

  def calcular_preco_final(self):
    # Acréscimo fixo de 5% sobre o preço base para Livros
    return self._preco_base * 1.05

  def exibir_detalhes(self):
    print("\n" + "-=" * 20)
    print("Detalhes do Produto Livro:")
    print(f"  Nome: {self.get_nome()}")
    print(f"  Autor: {self._autor}")
    print(f"  Páginas: {self._paginas}")
    print(f"  Preço Base: R${self.get_preco_base():.2f}")
    print(f"  Quantidade em Estoque: {self.get_quantidade_estoque()}")
    print(f"  Preço Final Calculado: R${self.calcular_preco_final():.2f}")
    print("-=" * 20)

class Vestuario(Produto):
  def __init__(self, nome, preco_base, quantidade_estoque, tamanho):
    super().__init__(nome, preco_base, quantidade_estoque)
    self._tamanho = tamanho

  def calcular_preco_final(self):
    return self._preco_base *1.05

  def exibir_detalhes(self):
    print("\n" + "-=" * 20)
    print("detalhes do produto vestuario: ")
    print(f"nome: {self.get_nome()}")
    print(f"preco base: {self.get_preco_base()}")
    print(f"quantidade no estoque: {self.get_quantidade_estoque()}")
    print(f"preco final calculado: {self.calcular_preco_final():.2f}")

lista_produtos = [] # Lista vazia para armazenar objetos Produto
dicionario_produtos = {} # Dicionário vazio para armazenar produtos por nome


eletronico01 = Eletronico("notebook", 2000, 5, 14)
eletronico02 = Eletronico("smart tv", 1500, 17, 24)

livro01 = Livro("programacao", 50, 20, "erik", 200)
livro02 = Livro("poo", 60, 40, "xavier", 350)

vestuario01= Vestuario('user34', 56, 34, 'm')
vestuario02 = Vestuario("anima14", 76, 54, 'm')

# Adicionar instâncias à lista_produtos
lista_produtos.append(eletronico01)
lista_produtos.append(eletronico02)
lista_produtos.append(livro01)
lista_produtos.append(livro02)
lista_produtos.append(vestuario01)
lista_produtos.append(vestuario02)

# Adicionar instâncias ao dicionario_produtos
dicionario_produtos[eletronico01.get_nome()] = eletronico01
dicionario_produtos[eletronico02.get_nome()] = eletronico02
dicionario_produtos[livro01.get_nome()] = livro01
dicionario_produtos[livro02.get_nome()] = livro02
dicionario_produtos[vestuario01.get_nome()] = vestuario01
dicionario_produtos[vestuario02.get_nome()] = vestuario02

print(f"total de produtos: {len(lista_produtos)}")
print(f'nome dos produtos: {list(dicionario_produtos.keys())}')


produto_em_produto = "notebook" # o produto existente
produto_em_produto = "tv box" # o produto inexistente
if produto_em_produto in dicionario_produtos:
  produto_encontrado0 = dicionario_produtos[produto_em_produto]
  print("produto encontrado")
  produto_encontrado0.exibir_detalhes()
else:
  print(f" o produto: {produto_em_produto} nao foi encontrado")

while True:
  print("oque deseja fazer?")
  print("-=" *30)
  print("1-> ver os produtos")
  print("2 -> adicionar um produto")
  print("3 -> buscar por um produto")
  print("4 -> excluir um produto")
  print("5 -> sair")
  usuario = int(input(": "))
  print()
  if usuario == 1:
    print("===detalhes de todos os produtos===")
    for produto in lista_produtos:
      produto.exibir_detalhes()
      print()

  if usuario == 2:
    print("1 -> livro")
    print("2 -> eletronico")
    print("3 -> vestuario")
    tipo_produto =int(input("tipo do produto: "))
    if tipo_produto == 1:

      try:
        nome_livro = str(input("nome do livro a ser adicionado: "))
        autor_livro = str(input("autor do livro: "))
        numero_paginas = int(input("numero de paginas: "))
        preco_livro = int(input("preco do livro: "))
        quantidade_livro = int(input("quantidade do livro:"))
        novo_livro = Livro(nome_livro, preco_livro, quantidade_livro, autor_livro, numero_paginas, )
        lista_produtos.append(novo_livro)
        dicionario_produtos[novo_livro.get_nome()] = novo_livro
        novo_livro.exibir_detalhes()
      except ValueError:
        print("erro no valor digitado")

    elif tipo_produto == 2:
      try:
        nome_eletronico = str(input("nome do eletronico: "))
        preco_eletronico = int(input(f"preco do(a) {nome_eletronico}: "))
        quantidade_eletronico = int(input(f"quantidade do(a) {nome_eletronico}: "))
        garantia_eletronico = int(input(f"garantia do(a) {nome_eletronico}:"))
        novo_eletronico = Eletronico(nome_eletronico, preco_eletronico, quantidade_eletronico, garantia_eletronico)
        lista_produtos.append(novo_eletronico)
        dicionario_produtos[novo_eletronico.get_nome()] = novo_eletronico
        novo_eletronico.exibir_detalhes()
      except ValueError:
        print(" erro nos dados digitados")

    elif tipo_produto == 3:
      nome_vestuario= str(input("nome do vestuario: "))
      quantidade_vestuario = int(input("quantidade: "))
      preco_vestuario = int(input("preco: "))
      tamanho_vestuario = int(input("tamanho: "))
      novo_vestuario = Vestuario(nome_vestuario, preco_vestuario, quantidade_vestuario, tamanho_vestuario)
      lista_produtos.append(novo_vestuario)
      dicionario_produtos[novo_vestuario.get_nome()] = novo_vestuario
    else:
      print("desculpe mais so pode adicionar livro e eletronico :(")
      print()

  elif usuario == 3:
    nome_produto_buscar = str(input("nome do produto que quer buscar: "))
    if nome_produto_buscar in dicionario_produtos:
      produto_encontrado = dicionario_produtos[nome_produto_buscar]
      produto_encontrado.exibir_detalhes()

    else:
      print(f" o produto  {nome_produto_buscar} nao estana lista de produtos")

  elif usuario == 4:
    produto_excluir = str(input("nome do produto para ser exluido: "))
    if produto_excluir not in dicionario_produtos:
      print("o produto nao existe! ")
    else:

      produto_a_remover = dicionario_produtos.pop(produto_excluir)

      for i, produto in enumerate(lista_produtos):
          if produto.get_nome() == produto_a_remover.get_nome():
              lista_produtos.pop(i)
              break
      print(f"Produto '{produto_excluir}' removido com sucesso.")

  elif usuario >= 6:
    print("opcao invalida")
    print()

  else:
    break
```
